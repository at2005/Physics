```python
import numpy as np

def exponentiate_hamiltonian(H, t):
    PLANCK_BAR = 6.62607015e-34 / (2 * np.pi)
    assert (hamiltonian == hamiltonian.T.conj()).all()

    eigenvalues, eigenvectors = np.linalg.eigh(H)
    sum_ops = 0
    for i, vec in enumerate(eigenvectors):
        sum_ops += np.exp((eigenvalues[i] * -1j * t) / PLANCK_BAR) * np.outer(vec, vec.conj()) 
    return sum_ops


pauli_y = np.array([[0, -1j], [1j, 0]])
pauli_x = np.array([[0, 1], [1, 0]])
pauli_z = np.array([[1, 0], [0, -1]])

hamiltonian = 32 * pauli_y + 12 * pauli_x + 43 * pauli_z

initial_state = np.array([1/np.sqrt(2), -1/np.sqrt(2)])

lst_vectors = []
for t in range(1, 5000):
    vec = (exponentiate_hamiltonian(hamiltonian, t) @ initial_state)
    lst_vectors.append(vec)
```

```python
# plot animation
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()
point, = ax.plot([], [], 'ro')
ax.set_ylim(-1, 1)
ax.set_xlim(-1, 1)

def init():
    point.set_data([], [])
    return point,

def animate(i):
    plt.cla()
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)

    x = np.real(lst_vectors[i])
    y = np.imag(lst_vectors[i])
    point.set_data(x, y)
    return point,

ani = animation.FuncAnimation(fig, animate, frames=len(lst_vectors), init_func=init, interval=50, blit=False)
plt.show()

# save animation
ani.save('animation.gif', writer='imagemagick', fps=24)
```
